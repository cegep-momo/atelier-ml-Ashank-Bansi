```python
# import les modules nécessaires
import numpy as nmp
import pandas as pd
```

```python
# configurer pandas pour afficher toutes les colonnes et éviter les avertissements de rétrocompatibilité
pd.set_option('display.max_columns', None) 
pd.set_option('future.no_silent_downcasting', True)
```

```python
# lire le fichier CSV contenant les données des Pokémons
nosPokemons = pd.read_csv("datas/pokedex.csv")
# afficher les noms des colonnes du DataFrame
print(nosPokemons.columns.values)
```

```python
print(nosPokemons.head(10))  # afficher les 10 premières lignes du DataFrame
```

```python
# nettoyer la colonne "LEGENDAIRE" en remplaçant les valeurs "VRAI" et "FAUX" par 1 et 0 respectivement, puis convertir la colonne en type entier
nosPokemonsClean = nosPokemons
nosPokemonsClean["LEGENDAIRE"] = nosPokemonsClean["LEGENDAIRE"].replace("VRAI", 1)
nosPokemonsClean["LEGENDAIRE"] = nosPokemonsClean["LEGENDAIRE"].replace("FAUX", 0)
nosPokemonsClean["LEGENDAIRE"] = nosPokemonsClean["LEGENDAIRE"].astype(int)
print(nosPokemonsClean.head())
```

```python
# afficher la forme et les informations du DataFrame nettoyé 
print(nosPokemonsClean.shape)
print(nosPokemonsClean.info())
```

```python
print(nosPokemonsClean[nosPokemonsClean['NOM'].isnull()])  # afficher les lignes où la colonne "NOM" est nulle
```

```python
# afficher les noms des Pokémons aux indices 61 et 63 pour retrouver les valeurs manquantes
print(nosPokemonsClean['NOM'][61])
print(nosPokemonsClean['NOM'][63])
```

```python
nosPokemonsClean['NOM'][62] = "Colossinge"  # remplacer la valeur manquante par "Colossinge"
# afficher le nom du Pokémon à l'indice 62, qui est manquant
print(nosPokemonsClean['NOM'][62])
```

```python
#Chargement des données des combats
combats = pd.read_csv("datas/combats.csv")

#Affichage des colonnes du Dataframe
print(combats.columns.values)

#Affichage des 10 premières lignes du Dataframe
print(combats.head(10))

#Comptage du nombre de lignes et de colonnes
print (combats.shape)

#Informations sur notre jeu de données
print (combats.info())
```

```python
#Calcul du nombre de fois où chaque Pokémon est en première position dans les combats
nbFoisPremierePosition = combats.groupby('Premier_Pokemon').count()
print(nbFoisPremierePosition)
```
